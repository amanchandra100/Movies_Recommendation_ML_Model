```python
import ast
# ast.literal_eval(string pas karna)  # int convert string object in integer obj
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c != 4:
            l.append(i['name'])
            c+=1
        else:
            break    
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_dir(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name']) 
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_dir)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
       y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('dancing')
```

```python
stem("captain barbossa, unique long believed to be dead, has come back to life and is headed to the edge of the earth with will turner and elizabeth swann. but nothing is quite as it seems. adventure fantas")
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
